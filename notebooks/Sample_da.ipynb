```python
df = []
for data in df_raw:
    # Find columns where all values are the same
    columns_to_delete = data.columns[data.nunique() == 1]
    
    # Drop those columns and assign it back to the DataFrame
    data.drop(columns=columns_to_delete, inplace=True)
    
    # Append the modified DataFrame to the new list
    df.append(data)
```
