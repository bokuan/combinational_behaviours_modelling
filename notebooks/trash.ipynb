```python
def find_lc_angle(df, minimum=45, maximum=80):
    matching_ptp_times = []

    i = 0
    while i < len(df):
        current_yaw = df.loc[i, 'angle']
        current_ptp_time = df.loc[i, 'ptp_time']
        found_matching_pair = False

        # Check the next maximum rows
        for j in range(i + 1, min(i + maximum + 1, len(df))):
            if (abs(df.loc[j, 'angle'] - current_yaw) <= 0.001) and \
                (abs(df.loc[round((i+j)/2), 'angle'] - current_yaw)>0.01) and \
                (abs(df.loc[round((i+j)/2), 'angle'] - current_yaw)<0.02):
                # Check if the matching row is within the range
                if j >= i + minimum:
                    matching_pair = (current_ptp_time.timestamp(), df.loc[j, 'ptp_time'].timestamp())
                    matching_ptp_times.append(matching_pair)
                    i = i + 41 
                    found_matching_pair = True
                    break  # If a match is found, exit the inner loop
                else:
                    i += 1
                    break

        if not found_matching_pair:
            i += 1
    print(matching_ptp_times)    
    return matching_ptp_times
```
